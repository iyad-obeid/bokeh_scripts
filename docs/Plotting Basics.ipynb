```python
import numpy as np
import numpy.fft as fft

# import matplotlib
# matplotlib.use('nbagg')
# matplotlib.use('Agg')
import matplotlib.pyplot as plt
pi = np.pi

fs = 1000
tmax = 2
t = np.array(range(tmax*fs))/fs
t.shape = (-1,1)
x = np.cos(2*pi*25*t)
XX = fft.fft(x,n=100000,axis=0)
# f = fs*np.array(range(XX.size))/XX.size
f = fft.fftfreq(n=100000)*fs

plt.figure(1)
plt.clf()
plt.plot(t,x)

plt.figure(2)
plt.clf()
plt.plot(f,np.abs(XX))
plt.xlim([0,fs/2])
plt.show()
```

```python
# prepare some data
x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]

# output to static HTML file
output_notebook()

# create a new plot with a title and axis labels
p = figure(title="simple line example", x_axis_label='x', y_axis_label='y')

# add a line renderer with legend and line thickness
p.line(x, y, legend="Temp.", line_width=2)

# show the results
show(p)
```
